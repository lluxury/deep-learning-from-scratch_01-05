### 神经网络
承上,感知机的权重是手动设置的神经网络的一个重要性质是它可以自动地从数据中学习到合适的权重参数  

$$
y = 
\begin{cases} 
   0,  b + x_1w_1 + x_2w_2 <= 0 \\
   1,  b + x_1w_1 + x_2w_2 > 0 
\end{cases}
$$
  
b 是被称为偏置的参数，用于控制神经元被激活的容易程度  
w1 和 w2 是表示各个信号的权重的参数，用于控制各个信号的重要性  
  
引进一个函数  
$$
y = h(b + w1x1 + w2x2) 
$$ 
$$ 
h(x) = 
\begin{cases} 
   0,  x <= 0 \\
   1,  x > 0 
\end{cases}
$$

#### 激活函数
a = b + w1x1 + w2x2  
y = h(a)  
神经元都是用一个○表示的, “神经元”和“节点”两个术语的含义相同, h(x) 激活函数,y输出  

##### sigmoid 函数
$$
h(x) = \sigma(x) = \frac{1}{1 + e^{-x}}
$$

```python
# 阶跃函数
def step_function(x):
    if x > 0:
        return 1
    else:
        return 0  
 # 只能接受实数 (浮点数) 作为输入,不能处理numpy的数组?        
```

```python
# coding: utf-8
import numpy as np
import matplotlib.pylab as plt


def step_function(x):
    return np.array(x > 0, dtype=int)
    # return np.array(x > 0, dtype=np.int)

X = np.arange(-5.0, 5.0, 0.1)
Y = step_function(X)
plt.plot(X, Y) # 绘制阶跃函数
plt.ylim(-0.1, 1.1)  # 指定图中绘制的y轴的范围
plt.show()
```

sigmoid 函数  
感知机中神经元之间流动的是 0 或 1 的二元信号，而神经网络中流动的是连续的实数值信号

```python
# coding: utf-8
import numpy as np
import matplotlib.pylab as plt


def sigmoid(x):
    return 1 / (1 + np.exp(-x))    

X = np.arange(-5.0, 5.0, 0.1)
Y = sigmoid(X)
plt.plot(X, Y)
plt.ylim(-0.1, 1.1)
plt.show()
```

ReLU函数
$$ 
h(x) = 
\begin{cases} 
   x,  x > 0 \\
   0,  x <= 0 
\end{cases}
$$

```python
# coding: utf-8
import numpy as np
import matplotlib.pylab as plt


def relu(x):
    return np.maximum(0, x)

x = np.arange(-5.0, 5.0, 0.1)
y = relu(x)
plt.plot(x, y)
plt.ylim(-1.0, 5.5)
plt.show()
```
